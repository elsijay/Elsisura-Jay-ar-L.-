```python
def hyphen(hamlet):
    jay=[]
    for i in range(len(hamlet)):
        if '-' in hamlet[i]:
            jay.append(hamlet[i-1]+'-')
    return jay
print(hyphen(hamlet))
```

```python
def ing(hamlet):
    jay=[]
    for i in range(len(hamlet)):
        if 'ing' in hamlet[i]:
            jay.append(hamlet[i-1]+'ing')
    return jay
print(ing(hamlet))
```

```python
def est(hamlet):
    jay=[]
    for i in range(len(hamlet)):
        if 'est' in hamlet[i]:
            jay.append(hamlet[i-1]+'est')
    return jay
print(est(hamlet))
```

```python
def punctuation(hamlet):
    jay=[]
    
    for i in range(len(hamlet)):
        if hamlet[i] in "!#$%&'()*-+./<=>@[\];<>,?{|}~":
            jay.append(hamlet[i-1]+hamlet[i])
    return jay

print(punctuation(hamlet))
```
